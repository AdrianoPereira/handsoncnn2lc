```python
import matplotlib.pyplot as plt
import xarray as xr
import geopandas as gpd
import os

import fiona
import rasterio
import rasterio.mask
```

```python
gdf = gpd.read_file('./data/test1/buffer05x05.shp')
all_classes = gdf['CLASSE PRI'].unique()
```

```python
for classe in all_classes:
    poly_sample = gdf[gdf['CLASSE PRI'] == classe].reset_index()
    
    OUTPUT = "./data/test1/buffer/%s/shp"%(classe)
    soja_filenames = list()
    for i, row in poly_sample.iterrows():
        temp = poly_sample.iloc[i:i+1, :]
        if temp.empty:
            continue
        path = os.path.join(OUTPUT, "%s"%(str(i).zfill(3)))
        print(path)
        if not os.path.exists(path):
            os.makedirs(path)
            os.makedirs(path.replace('shp', 'tif'))
        path = os.path.join(path, 'img%s.shp'%(str(i).zfill(3)))
        soja_filenames.append(path)
        temp.to_file(path)
        
    try:
        for file in soja_filenames:
            with fiona.open(file, "r") as shapefile:
                shapes = [feature["geometry"] for feature in shapefile]

            with rasterio.open("./data/raw/scene/REP_LO82200692018233.tif") as src:
                out_image, out_transform = rasterio.mask.mask(src, shapes, crop=True)
                out_meta = src.meta

            out_meta.update({"driver": "GTiff",
#                              "height": out_image.shape[1],
#                              "width": out_image.shape[2],
                             "height": 33, "width": 33,
                             "transform": out_transform
                            }
                           )

            with rasterio.open(file.replace('shp', 'tif'), "w", **out_meta) as dest:
                dest.write(out_image[:, :33, :33])
    except ValueError:
        print('Window not interset', classe)
```

```python
# soja_poly = gdf[gdf['CLASSE PRI'] == 'Area Urbana'].reset_index()
soja_poly = gdf[gdf['CLASSE PRI'] == 'Soja'].reset_index()
soja_poly.count()
```

```python
# OUTPUT = "./bahia/area_urbana/shp"
OUTPUT = "./bahia/buffer/soja/shp"
soja_filenames = list()
for i, row in soja_poly.iterrows():
    temp = soja_poly.iloc[i:i+1, :]
    if temp.empty:
        continue
    path = os.path.join(OUTPUT, "%s"%(str(i).zfill(3)))
    print(path)
    if not os.path.exists(path):
        os.makedirs(path)
        os.makedirs(path.replace('shp', 'tif'))
    path = os.path.join(path, 'img%s.shp'%(str(i).zfill(3)))
    soja_filenames.append(path)
    temp.to_file(path)
```

```python
for file in soja_filenames:
    with fiona.open(file, "r") as shapefile:
        shapes = [feature["geometry"] for feature in shapefile]

    with rasterio.open("./landsat_stack/bahia.tif") as src:
        out_image, out_transform = rasterio.mask.mask(src, shapes, crop=True)
        out_meta = src.meta

    out_meta.update({"driver": "GTiff",
                     "height": out_image.shape[1],
                     "width": out_image.shape[2],
#                      "height": 20, "width": 20,
                     "transform": out_transform
                    }
                   )

    with rasterio.open(file.replace('shp', 'tif'), "w", **out_meta) as dest:
        dest.write(out_image)
```
